```python
import numpy as np
from scipy.stats import linregress

# Perform linear regression and plot fit lines for each dataset
def plot_loss_of_velocity_with_linear_fit(glass_thickness, loss_velocity_data, label):
    plt.scatter(glass_thickness, loss_velocity_data, label=label)
    
    # Perform linear regression
    slope, intercept, r_value, p_value, std_err = linregress(glass_thickness, loss_velocity_data)
    
    # Create x values for the fit line
    x_fit = np.array([min(glass_thickness), max(glass_thickness)])
    y_fit = intercept + slope * x_fit
    
    # Plot the fit line
    plt.plot(x_fit, y_fit, linestyle='--', color='gray', label=f'Fit Line ({slope:.2f})')

# Plotting each firearm's data with linear fit lines
plt.figure(figsize=(10, 6))  # Increase figure size for clarity
plot_loss_of_velocity_with_linear_fit(glass_thickness, loss_velocity_22_rifle, '.22 Rifle')
plot_loss_of_velocity_with_linear_fit(glass_thickness, loss_velocity_303_rifle, '.303 Rifle')
plot_loss_of_velocity_with_linear_fit(glass_thickness, loss_velocity_38_revolver, '.38 Revolver')
plot_loss_of_velocity_with_linear_fit(glass_thickness, loss_velocity_9mm_pistol, '9mm Pistol')
plot_loss_of_velocity_with_linear_fit(glass_thickness, loss_velocity_45_pistol, '.45 Pistol')

# Adding titles and labels
plt.title('Loss of Velocity with Linear Fit Lines by Glass Thickness')
plt.xlabel('Glass Thickness (mm)')
plt.ylabel('Loss of Velocity (m/sec)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Data for plotting
glass_thickness = [2, 3, 5.5]  # in mm

# Loss of Velocity data for each firearm (m/sec)
loss_velocity_38_revolver = [12, 20, 45]
loss_velocity_9mm_pistol = [7, 15, 51]
loss_velocity_45_pistol = [18, 15, 39]

# Combine all the data into a single array
all_thickness = np.array(glass_thickness * 3)
all_loss_velocity = np.array(loss_velocity_38_revolver + loss_velocity_9mm_pistol + loss_velocity_45_pistol)

# Fit a linear regression line to the combined data
slope, intercept, r_value, p_value, std_err = linregress(all_thickness, all_loss_velocity)

# Create a function for the linear fit
def linear_fit(x):
    return slope * x + intercept

# Calculate the predicted values using the linear fit
predicted_values = linear_fit(all_thickness)

# Plot the data points and the linear fit for the combined data
plt.scatter(all_thickness, all_loss_velocity, label='Data')
plt.plot(all_thickness, linear_fit(all_thickness), color='red', label='Linear Fit')
plt.xlabel('Glass Thickness (mm)')
plt.ylabel('Loss of Velocity (m/sec)')
plt.legend()
plt.title('Linear Fit for Loss of Velocity vs. Glass Thickness (Combined Data)')
plt.grid(True)
plt.show()

# Print the slope and intercept of the linear fit
print(f'Slope: {slope}')
print(f'Intercept: {intercept}')
print(f'R-squared value: {r_value**2}')
```

```python
# Define a function to calculate the mean squared error (MSE)
def mean_squared_error(observed, predicted):
    return np.mean((observed - predicted)**2)

# Calculate the mean squared error for the fit
mse = mean_squared_error(all_loss_velocity, predicted_values)

# Print the mean squared error
print(f'Mean Squared Error (MSE) of the Quadratic Fit: {mse}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Data for plotting
# The added 0 represents no glass, indicating no loss of velocity.
glass_thickness = [0, 2, 3, 5.5]  # in mm
loss_velocity_38_revolver = [0, 12, 20, 45]
loss_velocity_9mm_pistol = [0, 7, 15, 51]
loss_velocity_45_pistol = [0, 18, 15, 39]

# Combine all the data
all_thickness = np.array(glass_thickness * 3)
all_loss_velocity = np.array(loss_velocity_38_revolver + loss_velocity_9mm_pistol + loss_velocity_45_pistol)

# Define a quadratic function
def quadratic_function(x, a, b, c):
    return a * x**2 + b * x + c

# Fit a quadratic curve to the combined data
params, _ = curve_fit(quadratic_function, all_thickness, all_loss_velocity)

# Calculate the predicted values using the quadratic fit
predicted_values = quadratic_fit(np.array(all_thickness), *params)

# Generate x values for the curve
x_values = np.linspace(min(all_thickness), max(all_thickness), 100)

# Calculate the corresponding y values using the quadratic function
y_values = quadratic_function(x_values, *params)

# Create a scatter plot of the combined data
plt.scatter(all_thickness, all_loss_velocity, label='Data')
# Plot the quadratic function curve
plt.plot(x_values, y_values, color='red', label='Quadratic Fit')
plt.xlabel('Glass Thickness (mm)')
plt.ylabel('Loss of Velocity (m/sec)')
plt.title('Quadratic Fit for Loss of Velocity vs. Glass Thickness (Combined Data)')
plt.legend()
plt.grid(True)
plt.show()

# Print the parameters of the quadratic fit
print(f'Quadratic Fit Parameters: a={params[0]}, b={params[1]}, c={params[2]}')
```

```python
# Define a function to calculate the mean squared error (MSE)
def mean_squared_error(observed, predicted):
    return np.mean((observed - predicted)**2)

# Calculate the mean squared error for the fit
mse = mean_squared_error(all_loss_velocity, predicted_values)

# Print the mean squared error
print(f'Mean Squared Error (MSE) of the Quadratic Fit: {mse}')
```
